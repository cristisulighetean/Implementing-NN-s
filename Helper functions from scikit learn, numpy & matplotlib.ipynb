## Description of np.meshgrid with examples

Return coordinate matrices from coordinate vectors.

Make N-D coordinate arrays for vectorized evaluations of N-D scalar/vector fields over N-D grids, given one-dimensional coordinate arrays x1, x2,â€¦, xn.

```python
# Description and examples of meshgrid

nx, ny = (3,2)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x, y)

print(xv)
print(yv)
```

```python
xv, yv = np.meshgrid(x, y, sparse=True)
print(xv, yv)
```

```python
x = np.arange(-12, 5, 0.1)
y = np.arange(-12, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=False)
z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)

h = plt.contourf(x,y,z)
plt.show()
```

## Numpy sum

```python
# Try out np.sum with different options

a = np.random.randint(low=1, high=5, size=(3,3))

# initial matrix
print(a)

sum1 = np.sum(a, axis=1, keepdims=True)
print("Sum with axis=1 -> row sum")
print(str(sum1))

sum2 = np.sum(a, axis=0, keepdims=True)
print("Sum with axis=0 -> column sum")
print(sum2)
```

## Np.argmax

```python
# np.argmax - returns the indices of the maximum values along an axis
# axis - 0 -> column axis
# axis - 1 -> row axis
b = np.arange(6)

np.argmax(b)  # Only the first occurrence is returned.
```
